```python
import numpy as np
import math
import h5py
import sys
import matplotlib.pyplot as plt
from sklearn import svm
from mlxtend.plotting import plot_decision_regions
from scipy import stats
```

```python
f1 = h5py.File('data_1.h5','r+') 
list(f1.keys())
X1 = f1['x']
y1=f1['y']
df1= np.array(X1[()])
dfy1= np.array(y1[()])
# print(df1,dfy1)
x0 = []
y0 = []
x1 = []
y1 = []
for i in range(len(df1)):
    if dfy1[i]==0:
        x0.append(df1[i][0])
        y0.append(df1[i][1])
    else:
        x1.append(df1[i][0])
        y1.append(df1[i][1])
model1 = svm.SVC(kernel = 'poly',degree=2,gamma='scale')
model1.fit(df1,dfy1)
plot_decision_regions(X=df1,y=dfy1,clf=model1)
```

```python
f1 = h5py.File('data_2.h5','r+') 
list(f1.keys())
X1 = f1['x']
y1=f1['y']
df1= np.array(X1[()])
dfy1= np.array(y1[()])
# print(df1,dfy1)
x0 = []
y0 = []
x1 = []
y1 = []
for i in range(len(df1)):
    if dfy1[i]==0:
        x0.append(df1[i][0])
        y0.append(df1[i][1])
    else:
        x1.append(df1[i][0])
        y1.append(df1[i][1])
def kernel2(x,y):
    ans = ((x@y.T)+1)**3
    return ans

model2 = svm.SVC(kernel = kernel2)
model2.fit(df1,dfy1)
plot_decision_regions(X=df1,y=dfy1,clf=model2)
```
